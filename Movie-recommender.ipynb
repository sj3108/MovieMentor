```python
import numpy as np
import pandas as pd
```

```python
movies= pd.read_csv('./data/tmdb_5000_movies.csv')
```

```python
credits=pd.read_csv('./data/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
merged_df = pd.merge(credits, movies, on='title', how='inner')
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
movies_main=merged_df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_main.info()
```

```python
movies_main.head()
```

```python
movies_main.isnull().sum()
```

```python
movies_main.dropna(inplace=True)
```

```python
movies_main.duplicated().sum()
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_main['genres']=movies_main['genres'].apply(convert)
```

```python
movies_main['keywords']=movies_main['keywords'].apply(convert)
```

```python
movies_main.head()
```

```python

def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_main['cast']=movies_main['cast'].apply(convert)
```

```python
movies_main.head()
```

```python
movies_main['cast'] = movies_main['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_main['crew'] = movies_main['crew'].apply(fetch_director)
```

```python
movies_main['overview']=movies_main['overview'].apply(lambda x:x.split())
```

```python
movies_main.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_main['cast'] = movies_main['cast'].apply(collapse)
movies_main['crew'] = movies_main['crew'].apply(collapse)
movies_main['genres'] = movies_main['genres'].apply(collapse)
movies_main['keywords'] = movies_main['keywords'].apply(collapse)
```

```python
movies_main.head()
```

```python
movies_main['tags']=movies_main['overview']+movies_main['genres']+movies_main['keywords']+movies_main['cast']+movies_main['crew']
```

```python
movies_main.head()
```

```python
new_df=movies_main[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Gandhi')
```
